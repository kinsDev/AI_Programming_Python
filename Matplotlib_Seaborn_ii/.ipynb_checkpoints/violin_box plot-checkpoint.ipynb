```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

fuel_econ = pd.read_csv('fuel_econ.csv')
fuel_econ.head()
```

```python
base_color = sb.color_palette()[0]

vehicle_sizes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
ordered_sizes = pd.api.types.CategoricalDtype(ordered = True, categories = vehicle_sizes)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(ordered_sizes)

sb.violinplot(data = fuel_econ, x = 'VClass', y = 'displ', color = base_color)
plt.xticks(rotation = 15)
```
